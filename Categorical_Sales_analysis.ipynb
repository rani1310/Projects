```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Exploratory Data Analysis**


**Gender**

```python
#to get a count of purchase by gender we use seaborn library

ax= sns.countplot (x='Gender',data=df)
```

```python
#since there isnt much information therefore we use label
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

**Age**

```python
#to analyze the data by age for each gender
ax = sns.countplot(x='Age Group', hue='Gender',data=df)
```

```python
#for labels in bar
ax = sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Analyze the amount spent by age group
ax = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot (x='Age Group',y='Amount',data=ax)
```

```python
#Total orders by state
ax = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot (x='State',y='Orders',data=ax)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Adjust rotation, horizontal alignment, and font size as needed
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#to view plot by maritak status
ax= sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#to analyze the amount spent based on gender for. marital status
ax = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=ax)
```

```python
#Lets check amount spent by Occupation
ax = sns.countplot(data=df, x='Occupation')

for bars in ax.containers:
      ax.bar_label(bars)

plt.xticks(rotation=45, ha='right', fontsize=10)  # Adjust rotation, horizontal alignment, and font size as needed
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Amount spent by Occupation

ax = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data=df,x='Occupation',y='Amount')

plt.xticks(rotation=45, ha='right', fontsize=10)  # Adjust rotation, horizontal alignment, and font size as needed
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Now, lets analyze amoubt spent by product category

ax = sns.countplot(data=df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.xticks(rotation=45, ha='right', fontsize=10)  #to adujust spacing between values
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Amount spent by product category
ax = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Product_Category',y='Amount',data=ax)
plt.xticks(rotation=45, ha='right', fontsize=10)  #to adujust spacing between values
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Top 10 Product ordered by Product Id

ax=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=ax,x= 'Product_ID',y='Orders')

```
