## Credit Card Fraud Detection using Random Forest


```python
## Import all libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score,roc_auc_score
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('/content/creditcard_2023.csv')
```

```python
Data.info()
```

```python
###To check if we are having any null values
Data.isnull().sum()
```

```python
##Since id is unique identifier so we will drop it its not required in features and targets variables .
##Also, since we need to predict class thats y we cant use class as our target variable
X= Data.drop(['id'], axis=1,errors='ignore')
Y= Data['Class']
```

```python
##To check column list we will print values

print(X.columns.tolist())
```

```python
###Divide the data into training and testing with random state

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
###To view training and testing value we use shape function
X_train.shape
```

```python
X_test.shape
```

```python
##Feature scaling is imporatnt in ML so that all features contribute equally to model process without dominating the other with less/sensitive values.

sacler = StandardScaler()
X_train_scaled = sacler.fit_transform(X_train)
X_test = sacler.transform(X_test)
```

```python
###Lets check class distribution in our training data sets
pd.Series(Y_train)
print(pd.Series (Y_train).value_counts(normalize=True))
```

```python
##Train the model for random model calssifier
rf_model = RandomForestClassifier(
n_estimators = 100,
max_depth = 10,
min_samples_split= 5,
random_state=42
)
```

```python
##We will trsin our model based on cvscores
cv_scores= cross_val_score(rf_model, X_train_scaled , Y_train, cv=5, scoring='f1')
print("\nCross-Validation f1 Scores:", cv_scores)
print("Average F1 score:", np.mean(cv_scores))
```

```python
rf_model.fit(X_train_scaled, Y_train)
```

```python
###Lets build a predictive model
X_prediction = rf_model.predict(X_test)
Y_prediction = rf_model.predict(X_train_scaled)
```

```python
##LETS Print the classification report
print(classification_report(Y_test, X_prediction))####0 ->non fraud, 1->fraud
```

```python
##Lets visualize confusion matrix through heatmap
plt.figure(figsize=(10,8))
cm = confusion_matrix(Y_test, X_prediction)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')###format annotation as int
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
####56750->true positive as we predidcted them to be non fraud, 56976->true negative as we predicted as fruad and it came to be fraud
```

```python
##lets analyze feature importance to reduce the impurity in the model .Higher the FI less is the impurity
importance = rf_model.feature_importances_
##feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': feature_importance})
feature_importance= pd.DataFrame({'Feature': X.columns, 'Importance': feature_importance}).sort_values('Importance', ascending=False)
print(feature_importance)
```

```python
feature_importance.head()
```

```python
###Now we will build feature imporatnce ranking
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
plt.title('Feature Importance Ranking')
plt.xlabel('Importance')
plt.ylabel('Feature')
```

```python
###As a next step we will be building fetaure correlation using confusin matrix
##feature correlation is imp to see how featuress are correlated .if value is 0.8 and above for 2 features then we need to exclude one to avoid instatbility due bto redundancy.
plt.figure(figsize=(10, 8))
cm = confusion_matrix = X.corr()
sns.heatmap(cm, annot=True, fmt='.2f', cmap='coolwarm')
##plt.xlabel('Predicted')
##plt.ylabel('Actual')
plt.tight_layout()
plt.title('Feature Corelation Matrix')
plt.show()
```

```python
##Lets looks model performance using RSC curve and AU score
##Lets looks model performance using RSC curve and AU score
from sklearn.metrics import roc_curve, roc_auc_score # Import roc_curve and roc_auc_score
y_pred_proba = rf_model.predict_proba(X_test)[:, 1]
fpr, tpr ,_ = roc_curve(Y_test, y_pred_proba)
roc_auc = roc_auc_score(Y_test, y_pred_proba) # Calculate AUC correctly
```

```python
###Now we will be building ROC Curve
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})') # Changed 'colors' to 'color'
plt.plot([0, 1], [0, 1], color='navy', linestyle='--') # Changed 'colors' to 'color'
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
