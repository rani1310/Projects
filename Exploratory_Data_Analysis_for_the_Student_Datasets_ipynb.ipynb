```python
# First, standardize your original Gender column
Data['Gender'] = Data['Gender'].str.strip().str.lower()

# Then map
gender_map = {'male': 'Male', 'female': 'Female'}
Data['Gender_label'] = Data['Gender'].map(gender_map)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=Data, x='Marks[10th]', y='Marks[12th]', hue='Gender_label')
```

```python
# If Gender was converted like {'Male': 0, 'Female': 1}, reverse it:
gender_map = {0: 'Male', 1: 'Female'}

# Create a temporary column just for plotting
Data['Gender'] = Data['Gender'].map(gender_map)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=Data, x='Marks[10th]', y='Marks[12th]', hue='Gender')
```
